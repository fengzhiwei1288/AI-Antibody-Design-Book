```python
import os

# Define the folder to save the .fas files
output_folder = "CD276-Negative-Sequences"

# Create the folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Read the sequences from the file
with open('sequence-active.txt', 'r') as file:
    content = file.read()

# Split the content into individual sequences based on the ">" identifier
sequences = content.split(">")

# Process each sequence (ignoring empty lines)
for seq in sequences:
    if seq.strip():  # check if it's not empty
        lines = seq.splitlines()
        header = lines[0].strip()  # the sequence identifier (e.g., "1")
        sequence = ''.join(lines[1:]).replace(" ", "")  # join lines and remove spaces
        
        # Define the output file path with the header as the filename
        output_file_path = os.path.join(output_folder, f"{header}.fas")
        
        # Save the sequence to the specified folder
        with open(output_file_path, 'w') as output_file:
            output_file.write(f">{header}\n{sequence}\n")

print("Sequences have been saved to individual .fas files in the folder:", output_folder)
```

```python
import os

# Define the path to the main directory and the output file
main_dir = "CD276-Negative-Sequences"
output_file = "job.lst-test"

# Open the output file in write mode
with open(output_file, "w") as f:
    # Write the header line
    f.write("##Target(components) Size(AAs) Name(for output)\n")
    
    # Loop through each folder in the main directory
    for folder_name in os.listdir(main_dir):
        # Ensure it's a folder and not a file
        folder_path = os.path.join(main_dir, folder_name)
        if os.path.isdir(folder_path):
            # Write the formatted line to the output file
            line = f"{folder_name}/CD276 1000 {folder_name}_CD276\n"
            f.write(line)

print(f"'job.lst' has been created with entries for each folder in '{main_dir}'.")
```

```python
import os

# Define the folder to save the .fas files
# 定义保存 .fas 文件的文件夹
output_folder = "CD276-Negative-Sequences1"

# Create the folder if it doesn't exist
# 如果文件夹不存在，则创建该文件夹
os.makedirs(output_folder, exist_ok=True)

# Read the sequences from the file
# 从文件中读取序列
with open('sequence-active.txt', 'r') as file:
    content = file.read()

# Split the content into individual sequences based on the ">" identifier
# 根据 ">" 标识符将内容分割成单个序列
sequences = content.split(">")

# Process each sequence (ignoring empty lines)
# 处理每个序列（忽略空行）
for seq in sequences:
    if seq.strip():  # check if it's not empty
        # 检查是否为空，避免处理空行
        lines = seq.splitlines()
        header = lines[0].strip()  # the sequence identifier (e.g., "1")
        # 序列标识符（例如 "1"）
        sequence = ''.join(lines[1:]).replace(" ", "")  # join lines and remove spaces
        # 连接序列行并去除空格
        
        # Define the output file path with the header as the filename
        # 使用标识符作为文件名定义输出文件路径
        output_file_path = os.path.join(output_folder, f"{header}.fas")
        
        # Save the sequence to the specified folder
        # 将序列保存到指定文件夹
        with open(output_file_path, 'w') as output_file:
            output_file.write(f">{header}\n{sequence}\n")
 
# Print confirmation message
# 打印确认信息
print("Sequences have been saved to individual .fas files in the folder:", output_folder)
```
